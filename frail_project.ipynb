```python
import pandas as pd
import numpy as np 
df = pd.read_csv('/Users/evangelosgeraridis/Desktop/ddcdm/big data/PROJECT Μεγαλοοικονόμου/clinical_dataset.csv', sep=';')
with pd.option_context('display.max_rows', None ):
    print(df.shape)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cols = ['gender', 'ortho_hypotension', 'weight_loss', 'gait_speed_slower','grip_strength_abnormal','low_physical_activity','memory_complain','living_alone','leisure_club','house_suitable_participant','house_suitable_professional','gait_optional_binary'  ]
df[cols] = df[cols].apply(LabelEncoder().fit_transform)

# όσες τιμές ήταν δύτιμες έγιναν με label encoder και όπου υπήρχε κενό έγινε 2 ( π.χ. yes->1 , no-> 0 , κενό >2)
df = df.replace({"Non frail": 0 ,
                        "Pre-frail": 1 , "Frail": 2 , "Sees moderately":1 , "Sees well": 0 , "Sees poorly" : 2 , "Hears well": 0 , "Hears moderately" : 1, "Hears poorly": 2, "No sleep problem" : 0,"Occasional sleep problem" : 1, "Permanent sleep problem" : 2 , "Never smoked":0 , "Past smoker (stopped at least 6 months)" : 1 , "Current smoker" :2}) # μέχρι εδώ έχουμε τις αρνητικές επιδράσεις στην υγιεία ,  more points -> worse health status
df = df.replace ({"<5 sec" : 0 , ">5 sec" : 1 , "1 - Very bad" : 1 , "2 - Bad" : 2 , "3 - Medium" : 3 , "4 - Good" :  4 , "5 - Excellent" : 5 ,"3 - About the same": 3 , "2 - A little worse" : 2, "4 - A little better": 4, "1 - A lot worse": 1, "5 - A lot better": 5 , "< 2 h per week" :1 , "> 2 h and < 5 h per week" : 2 , "> 5 h per week" : 3}) # more points -> better health 
 # more points-> better health status στο δεύτερο df.replace
df.activity_regular = df.activity_regular.replace ({'No' : 0})  # ειδικά στο  activity_regular  το No -> 0 , γιατί υπάρχουν και άλλα No στο df
       
 # τα δύο αυτά replaces μαζί με το τρίτο γίνoνται και πιο εύκολα με το label encoding της scikit , αλλά διαισθητικά έχει περισσότερο νόημα να φαίνεται η υγεία του ασθενούς διαβαθμισμέμη με το 0 υγιής , 1 χειρότερη ... και αντίστοιχα για την βελτίωση ή την επιδείνωση της υγιείας του   
    # PART A.1 done (with domain knowledge usage , if without sklearn label encoder can work too              
df = df.replace({ 999 : np.nan  ,  "test non realizable" : np.nan , " Test not adequate" : np.nan })
df = df.fillna(np.nan)                                                                      # obvious outliers και erroneous -> NaN
df.house_suitable_participant = df.house_suitable_participant.replace ({ 2 : np.nan})
df.house_suitable_professional = df.house_suitable_professional.replace ({ 2 : np.nan})  # ήταν δύτιμες και επειδή είχαν missing values , αυτά έγιναν 2 με το label encoding , οπότε τώρα τις κάνουμε NaN

print(df)
print(df.dtypes)
print(df.describe())     # statistical info mean , min values , max values , θα χρειαστούν για τα fillings μετά
print(df.info())         #check ότι όλα είναι ακόμα float μαζί με τις nan τιμές 
                                # δεν τελειώσαμε με τα outliers , καθώς πέραν της προφανούς τιμής 999 μπορεί να υπάρχουν και αρνητικές τιμές ή άλλες που δε μπορεί να είναι legit , τις οποίες καλό θα είναι να τις ελέγξουμε όχι κοιτάζοντας όλο το dataset , καθώς είναι μεγάλο
 # αφού βγάλαμε τα obvious outliers , τα 999 δηλαδή , τώρα κάνουμε zscore για τις υπόλοιπες στήλες με καθαρά αριθμητικά δεδομένα , έστω col_zscore η λίστα αυτή


# κοιτάζοντας το df.info() στα numeric features (αυτά που ήταν εξ'αρχής numeric ) δεν έχουμε πληροφορίες για τα  raise_chair_time , waist και σε τι κλίμακα μετρώνται,
# για να μπορούμε κάνοντας zscore να διώξουμε τα outliers αν υπάρχουν οπότε αυτά  δε θα συμμάσχουν στο zscore test. To ίδιο θα ισχύσει και  για 
# leisure_out , leasure_club, social_visits , social_calls , social_texts( δεν ξέρουμε για το πλήθος τους , μπορεί να είναι σύντομα )
# όλα τα self-rated features (δεν είναι καν μετρήσεις πραγματικές και το max τους δεν είναι εξωφρενικό) ,
# stairs_number (δεν ξέρουμε αν μένουν σε πολυόροφη πολυκατοικία ) ,  katz_index  , iadl_grade , comorbidities_count   (καθώς θεωρούμε ότι κάποιος ηληκιωμένος πιθανόν να έχει 17 παθήσεις ( αυτό είναι το μέγιστο στο df μας )) ,
# comorbidities_significant_count  , medication_count   (17 φάρμακα που είναι το μέγιστο είναι φυσιολογικό θεωρούμε)
# τα features falls_one_year και fractures _ three_years θεωρούνται επίσης φυσιολογικά καθώς το ένα έχει max 40 και το άλλο 4 (40 πτώσεις το χρόνο μπορεί να 
# συμβεί σε έναν ηληκιωμένο και κατάγματα το ίδιο .. 
#το gait_get_up έχει μέγιστο τα 60 δεύτερα μέχρι να πάει ο ηλικιωμένος άνθρωπος για να κάνει το test ( φυσιολογικό)

# τα screening_score , cognitive_total_score , mmse_total_score είναι test από ειδικούς δεν έχουμε domain knowledge για z score


# το gait_speed_4m έχει μέγιστο 45 δεύτερα , χρόνος για να διανύσει 4 μέτρα , το οποίο είναι πολύ (θα γίνει z score για να αφαιρεθούνα τα outliers)
# τα hospitalization_one_year και αντίστοιχα το 3 years . Ωστόσο εδώ οι τιμές έχουν μεγάλη απόκλιση , οπότε καλό θα ήταν να μην κάνουμε z score για τα outliers γιατί 
# είναι πολύ πιθανό να χάσουμε και τιμές που είναι φυσιλογικές ( όπως 4 νοσηλείες το έτος , φυσιολογικό όχι μόνο για ηληκιωμένους ) 

# Εν τέλει θα κάνουμε z score μόνο στην στο gait_speed_4m και στο bmi score ( καθώς αρχικά αυτά είναι πιο πιθανό να προέρχονται από κανονικά δεδομένα ) , ενώ τα 
# υπόλοιπα είτε είναι αρμοδιότητες των domain experts , είτε έχουν μεγάλη διασπορά ... Θα μπορούσαμε να κάνουμε και για την ηληκία z score , αλλά έχει 
# μέγιστη τιμή το 95 , που είναι σίγουτα εφικτή ηληκία...


df_zscore = df[['bmi_score' , 'gait_speed_4m']] #φτιάχνουμε ένα νέο dataframe για το zscore 
    
    
from scipy.stats import zscore
df1 = zscore(df_zscore,nan_policy='omit')
df_zscore = df[[ 'bmi_score' ,'gait_speed_4m']].where ( abs(df1) <3 , np.nan)
with pd.option_context('display.max_rows', None , 'display.max_columns', None): 
    print(df_zscore)
    print (df_zscore.info())
    

import seaborn as sns
sns.boxplot(df_zscore['bmi_score'])
df[['bmi_score','gait_speed_4m']] = df_zscore
sns.boxplot(df['bmi_score'])

# τώρα βλέπουμε καλύτερα αποτελέσματα καθώς 44 μέγιστο bmi score -> extreme obessity ( υπαρκτό ωστόσο ενώ το 921 σίγουρα όχι)
# για το χρόνο περπατήματος 4 μέτρων έχω τώρα ως μέγιστο τα 18 δεύτερα , που είναι επίσης πολύ , αλλά αρκετά καλύτερα από το 45 

df = df.where(df>=0 , np.nan) # διώχνουμε και τυχόν αρνητικές τιμές , καθώς όλα τα features έχουν μη αρνητικά values


# Ξεκινάει η manual διαδικασία μείωσης διαστατικότητας

# επειδή έχουμε πολλές στήλες (features ) , τα οποία σίγουρα έχουν μια εξάρτηση μεταξύ τους (για παράδειγμα hosp1year και hosp3years , προφανές)
# θα φτιάξουμε έναν πίνακα συσχετίσεων αυτών των μεταβλητών και έτσι θα δούμε ποια features έχουν μεγάλο συντελεστή συσχέτισης (r) για να 
# διώξουμε ένα από αυτά . Αν r κοντά στο 1 ή -1 μπορώ με πολύ μεγάλη ακρίβεια να προβλέψω τις τιμές ενός χαρακτηριστικού , δοθέντος ότι γνωρίζουμε το άλλο.

# φτιάχνουμε το dfc ( το οποίο είναι το ίδιο το df χωρίς όμως το fried feature ). Επειδή ο σκοπός της έρευνας εδώ είναι να χαρα/στεί ο 
# participant ως fried , δε θέλω να διώξω features συσχετιζόμενα με το fried , ή ακόμη χειρότερα και το ίδιο το fried 
from sklearn import datasets


dfc = df.drop(['fried'], axis=1)


# φτιάχνουμε τον correl πίνακα , ο οποίος θα περιέχει τις απόλυτες τιμές των r ( δε μας ενδιαφέρει θετική ή αρνητική συσχέτιση αρκεί να είναι κοντά στο 
# 1 ή -1 η αρνητική) . Στο συγκεκριμένο project θα θεωρήσουμε την συσχέτιση ως σημαντική αν είναι >=0.5 κατά απόλυτη τιμή 

correl = dfc.corr()

#γραφικά σαν heatmap
import matplotlib.pyplot as plt
import seaborn as sns
def correlation_heatmap(df):
    correlations = df.corr()

    fig, ax = plt.subplots(figsize=(80,80))
    sns.heatmap(correlations, vmax=1.0, center=0, fmt='.2f',
                square=True, linewidths=.9, annot=True, cbar_kws={"shrink": 0.7})
    plt.show();
    
correlation_heatmap(df)

# σαφώς δε με απασχολούν οι τιμές της διαγωνίου ,αφού αυτές είναι 1 ( είναι ο συντελεστής συσχέτισης του χαρ/κού με τον ευατό του, επομένως πλήρως θετική εξάρτηση

# άρα θα πάρουμε είτε τον άνω τριγωνικό , ή κάτω αντίστοιχα (στοιχεία πάνω ή κάτω της διαγωνίου ) (συντ συσχετ μεταξύ διαφορετικών χαρ/κών)

upper_correl =  correl.where(np.triu(np.ones(correl.shape),k=1).astype(bool)) # πήραμε το άνω τρίγωνο
to_drop = [column for column in upper_correl.columns if any(upper_correl[column] >= 0.5)]

# μόνο το άνω τρίγωνο πήραμε , καθώς το κάτω τρίγωνο καθρεπτίζει τα στοιχεία του πάνω (συσχετίσεις μεταξύ ιδίων features) , με άλλα λόγια το 
# μητρώο correl είναι συμμετρικό ως προς τη διαγώνιο, οπότε είτε πάρω άνω ή κάτω , τα ίδια αποτελέσματα θα πάρω πίσω

with pd.option_context('display.max_rows', None , 'display.max_columns', None): 
    print(to_drop)
    
# βλέπουμε ότι μπορούμε να διώξουμε τις εξής στήλες hospitalization_three_years', 'gait_speed_4m', 'gait_speed_slower', 'bmi_body_fat', 'waist', 
# 'lean_body_mass', 'mmse_total_score', 'house_suitable_professional', 'medication_count'
# καθώς το hosp1year με το hosp3year έχουν r=0.84 και οι υπόλοιπες με κάποια άλλη που θα μείνει στο dataframe 

# το gait_speed_slower χρησιμοποήθηκε για να γίνει η ταξινόμηση του fried , οπότε το κρατάμε προς το παρόν
c = df.corr()
with pd.option_context('display.max_rows', None , 'display.max_columns', None): 
    print(c.where(abs(c)>=0.35))

# φτιάξαμε ένα correlation matrix , το c εδώ για να ελέγξουμε πριν γίνει η ανάλυση , ποια features πλυν των 5 βασικών , έχουν μεγαλύερη συσχέτιση με το fried,
# αλλά και ποια features από το dataframe μας έχουν μεγάλη συσχέτιση με αυτά που προτείνεται να διώξουμε
# επειδή λοιπόν το gait_speed_4m έχει μεγάλη συσχέτιση με το gait_get_up ( με μόλις 7 περισσότερα missing values ) και αρκετά μεγαλύτερη συσχέτιση με το fried
# επιλέγουμε να διώξουμε το gait_get_up αντί του gait_speed_4m , τα υπόλοιπα ως έχουν

df = df.drop(['hospitalization_three_years','gait_speed_slower' , 'gait_get_up', 'bmi_body_fat', 'waist', 'lean_body_mass', 'mmse_total_score', 'house_suitable_professional', 'medication_count' ] , axis = 1)
  
with pd.option_context('display.max_rows', None , 'display.max_columns', None): 
    print(df)
    print(df.info())
    
# Πριν ξεκινήσουμε την διαδικασία filling, θα πάμε πρώτα να κοιτάξουμε αν κάποια features με πολλά missing values ,σίγουρα κάτω από 500 , ίσως και λίγα
# παραπάνω , πως σχετίζονται αυτά με την παράμετρο fried . Ο κύριος στόχος μετά είναι να ταξινομήσουμε ( classification ) , για να προβλέψουμε το fried
# οπότε αν κάτι έχει πολλά missing values και είναι δεν είναι χρήσιμο για την πρόβλεψη της παραμέτρου fried , τότε θα το διώξουμε
# δηλαδή ο συντελεστής συσχέτισης R να είναι μικρός σχεσιακά με το fried


with pd.option_context('display.max_rows', None , 'display.max_columns', None): 
    print(df.corr())

# για τον άνωθεν λόγο το feature house_suit_partic φεύγει

df = df.drop(['house_suitable_participant'] , axis = 1)


    
# Filling missing values 

# Στις στήλες τις οποίες δεν έχουν πολλά missing values το πολύ 10% λείπει , μπορούμε απλά να τις γεμίσουμε με την διάμμεσο των παρατηρήσεων της στήλης 
# διαχωρισμένες οι παρατηρήσεις αναλόγως με ποια κατηγορία fried ανήκουν χωρίς ιδιαίτερα μεγάλο σφάλμα
# παίρνουμε δηλαδή ένα κατηγορικό feature ( το πιο χαρακτηριστικό στο dataframe και χωρίς NaN values) και στη στήλη που έχει missing values ,
# υπολογίζεται η διάμμεσος των παρατηρήσεων αυτής της στήλης που ανήκουν στο frail , pre-frail και non frail καιανάλογα αν το missing value 
# ανήκει στο frail ,non frail , pre-frail , παίρνει σαν τιμή την αντίστοιχη διάμμεσο των παρατηρήσεων που ανήκουν σε αυτή την ομάδα
# Αυτό μπορούμε να το δούμε από το df.info() , κοιτάμε το πλήθος των non null



df_medfil = df[['social_phone','sleep' , 'falls_one_year' , 'gait_speed_4m'  , 'depression_total_score' , 'anxiety_perception' , 'leisure_out' , 'social_visits' , 'social_calls' , 'social_skype' , 'social_text' , 'life_quality' , 'health_rate' , 'health_rate_comparison' , 'pain_perception' , 'activity_regular' , 'smoking' , 'alcohol_units' , 'iadl_grade' , 'bmi_score' , 'fractures_three_years' , 'raise_chair_time' , 'balance_single' , 'cognitive_total_score' ]]
df_medfil = df_medfil.fillna(df_medfil.groupby(df['fried']).transform('median'))


df_medfil['fried'] = df['fried']   # βάζουμε πάλι το fried για να ελέγξουμε αν χάλασαν οι συσχετίσεις
c = df_medfil.corr() #ξαναφτιάχνουμε τον correlation matrix για να δούμε αν "χάλασε" κάποια συσχέτιση με το fried

with pd.option_context('display.max_rows', None , 'display.max_columns', None): 
    print(df_medfil)
    print(c)
    
# βλέπουμε ότι δε χάλασαν, για την ακρίβεια ο r σχετικά με το fried δεν άλλαξε καθόλου (στο  επίπεδο που θέλουμε ούτε στο δεύτερο δεκαδικό)
# οπότε πετάμε τις στήλες του medfil στις αντίστοιχες του αρχικού df 


df[['social_phone', 'sleep' , 'falls_one_year' , 'gait_speed_4m'  , 'depression_total_score' , 'anxiety_perception' , 'leisure_out' , 'social_visits' , 'social_calls' , 'social_skype' , 'social_text' , 'life_quality' , 'health_rate' , 'health_rate_comparison' , 'pain_perception' , 'activity_regular' , 'smoking' , 'alcohol_units' , 'iadl_grade' , 'bmi_score' , 'fractures_three_years' , 'raise_chair_time' , 'balance_single' , 'cognitive_total_score' ]] = df_medfil[['social_phone','sleep' , 'falls_one_year' , 'gait_speed_4m'  , 'depression_total_score' , 'anxiety_perception' , 'leisure_out' , 'social_visits' , 'social_calls' , 'social_skype' , 'social_text' , 'life_quality' , 'health_rate' , 'health_rate_comparison' , 'pain_perception' , 'activity_regular' , 'smoking' , 'alcohol_units' , 'iadl_grade' , 'bmi_score' , 'fractures_three_years' , 'raise_chair_time' , 'balance_single' , 'cognitive_total_score' ]] 


# για το classification , πετάμε έξω από το df όλες τις στήλες αυτές που χρησιμοποιήθηκαν στην ανάλυση του frail_safe , καθώς δε θα χρησιμοποιηθούν πουθενά 
# στην ανάλυσή μας

# εφόσον το stairs_number , έχει πολλές άγνωστες τιμές και δεν έχει κάποια σημαντική συσχέτιση με το fried θα φύγει επίσης

df = df.drop(['weight_loss', 'exhaustion_score', 'grip_strength_abnormal','low_physical_activity' , 'stairs_number'] , axis = 1)

print(df.info())
correl = df.corr() # νέος correlation matrix σύμφωνα με το πιο πρόσφατο df

with pd.option_context('display.max_rows', None , 'display.max_columns', None): 
    print(correl)
    
# CLASSIFICATION

# θα χρησιμοποιήσουμε μοντέλα της sklearn , πρώτα θα τα εκπαιδεύσουμε και μετά τα θα τα χρησιμοποιήσουμε για πρόβλεψη

# χωρίζουμε αρχικά το δείγμα στα features που θα χρησιμοποιηθούν για πρόβλεψη και έπειτα σε αυτό που θέλω να προβλέψω (fried)

# για πρόβλεψη θα χρησιμοποιήσω όσα σχετίζονται όσο το δυνατόν περισσότερο με το fried , δηλαδή όσα έχουν τον μεγαλύτερο συντελεστή συσχέτισης r (ελέγχοντας τον
# correlation matrix) . Ιδανικά θα ήθελα τα features να έχουν ισχυρή εξάρτηση με το  fried , αλλά αναμεταξύ τους να είχαν χαμηλή εξάρτηση (σχεδόν ανεξάρτητα)
# γιατί έτσι λογικά θα είχαμε τη βέλτιστη πρόβλεψη , ταξινόμηση του fried βάση αυτών
# ωστόσο δεν είναι πολλά αυτά τα χαρ/κά οπότε δε χρειάζεται και τέτοιος έλεγχος , παραμόνο εξάρτηση με το fried

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score   
# θα χρειαστούμε το accuracy_score, για να συγκρίνουμε τους διάφορους αλγόριθμους 

X = df[['activity_regular' , 'gait_speed_4m' , 'balance_single' , 'cognitive_total_score', 'depression_total_score' , 'age' , 'pain_perception' ,'health_rate' , 'leisure_out']]
y = df[['fried']]#τα Χ θα χρησιμοποιηθούν για predictions και το y τα predicted (χωρίζουμε το δείγμα μας ουσιαστικά σε 2 μέρη)
y = np.ravel(y,order='C')
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size = 0.2)
# Τα παρακάτω δεν τρέχονται , καθώς έχουμε ήδη αποθηκεύσει το μοντέλο μας ( logistic reggression )
# καθώς και τα xtest και ytest σε csv αρχεία γιατί με εκείνο το μοντέλο με εκείνη την εκαπίδευσή του πετύχαμε καλύτερα αποτελέσματα
#xt = X_test
#yt = y_test
#xtr = X_train 
#ytr = y_train 


#xt.to_csv(r'/Users/evangelosgeraridis/Desktop/ddcdm/big data/PROJECT Μεγαλοοικονόμου\xt.csv', index = False)
#yt.tofile('/Users/evangelosgeraridis/Desktop/ddcdm/big data/PROJECT Μεγαλοοικονόμου/yt.csv', sep = ',')
# τα X_train ,y_train για την εκπαίδευση του μοντέλου μας (machine learning) , 
xt = pd.read_csv(r'/Users/evangelosgeraridis/Desktop/ddcdm/big data/PROJECT Μεγαλοοικονόμου\xt.csv')
yt = np.genfromtxt("/Users/evangelosgeraridis/Desktop/ddcdm/big data/PROJECT Μεγαλοοικονόμου/yt.csv" , delimiter=",")
print(xt)
print(yt)
                                                       
# τα X_test , y_test για να ελέγξουμε τα αποτελέσματα μετά . Το 80% των παρατηρήσεων θα χρησιμοποιηθεί για train και το 20% για test 

from sklearn import model_selection
import pickle

# KNN

from sklearn.neighbors import KNeighborsClassifier              # Αρχικά δοκιμάζω με knn, καθώς δεν έχω πολλές μεταβλητές σχετικά ( δεν είναι κοντά στις 15 )
knn = KNeighborsClassifier()                                    
knn.fit(X_train, y_train)
predictions_knn = knn.predict(X_test)
score_knn = accuracy_score(y_test , predictions_knn)                                                       
print(score_knn)
                                                       
# βλέπω accuracy του test set 0.57 , καλό για αρχή

# θα δούμε τώρα και υπόλοιπους αλγορίθμους classification 

# DECISION TREES 

#X = df[['activity_regular' , 'gait_speed_4m' , 'balance_single' , 'cognitive_total_score', 'depression_total_score' , 'age' , 'pain_perception' ,'health_rate' , 'leisure_out']]
#y = df[['fried']]  
#y = np.ravel(y,order='C')
#X_train , X_test , y_train , y_test = train_test_split(X,y , test_size = 0.2)  # ξανακάνουμε train για τον δεύτερο αλγόριθμο όμοια και στους άλλους
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier().fit(X_train, y_train)
predictions_dt = clf.predict(X_test)
score_dt = accuracy_score(y_test , predictions_dt)                                                       
print(score_knn)
                               
# δεν αποθηκεύουμε πλέον άρα σε σχόλια 
#filename = 'finalized_modeldt.sav'
#pickle.dump(logreg, open(filename, 'wb'))

#score_dt = accuracy_score(y_test , predictions_dt)

#print(score_dt)                                                       

# LOGISTIC REGRESSION 

from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
predictions_LR = logreg.predict(X_test)
y_pred = predictions_LR
score_lr = accuracy_score(y_test , y_pred)                                                       
print(score_lr)
 
 #for time purposes the following section is viwed as comments , though the confusion matrix here is very helpful   
#from sklearn.metrics import plot_confusion_matrix
#plot_confusion_matrix(logreg, xt, yt)
#from sklearn.metrics import classification_report
#print(classification_report(yt,y_pred))
#score = logreg.score(xt, yt)
#print(score)



# Gaussian Naive Bayes

#X = df[[ 'activity_regular' , 'gait_speed_4m' , 'balance_single' , 'cognitive_total_score', 'depression_total_score' , 'age' , 'pain_perception' ,'health_rate' , 'leisure_out']]
#y = df[['fried']]
#y = np.ravel(y,order='C')
#X_train , X_test , y_train , y_test = train_test_split(X,y , test_size = 0.2)  
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
predictions_gnb = gnb.predict(X_test)
#filename = 'finalized_modelgnb.sav'
#pickle.dump(clt, open(filename, 'wb'))
score_gnb = accuracy_score(y_test , predictions_gnb) 
print(score_gnb) 
#from sklearn.metrics import classification_report
#print(classification_report(y_test,y_pred))



# RANDOM FOREST CLASSIFICATION 

#X = df[[ 'activity_regular' , 'gait_speed_4m' , 'balance_single' , 'cognitive_total_score', 'depression_total_score' , 'age' , 'pain_perception' ,'health_rate' , 'leisure_out']]
#y = df[['fried']]
#y = np.ravel(y,order='C')
#X_train , X_test , y_train , y_test = train_test_split(X,y , test_size = 0.1)                                                        
from sklearn.ensemble import RandomForestClassifier
rfclf=RandomForestClassifier()  
rfclf.fit(X_train,y_train)
y_pred=rfclf.predict(X_test)
predictions_rfclf = rfclf.predict(X_test)
#filename = 'finalized_modelrcclf1.sav'
#pickle.dump(rfclf, open(filename, 'wb'))
score_rfclf = accuracy_score(y_test , predictions_rfclf) 
print(score_rfclf)
#from sklearn.metrics import classification_report
#print(classification_report(y_test,y_pred))
#import matplotlib.pyplot as plt
#import seaborn as sns
#class_names=[0,1,2] # name  of classes
#fig, ax = plt.subplots()
#tick_marks = np.arange(len(class_names))
#plt.xticks(tick_marks, class_names)
#plt.yticks(tick_marks, class_names)
# create heatmap
#sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
#ax.xaxis.set_label_position("top")
#plt.tight_layout()
#plt.title('Confusion matrix', y=1.2)
#plt.ylabel('Actual label')
#plt.xlabel('Predicted label')
#from sklearn.metrics import classification_report
#print(classification_report(y_test,y_pred))
#from sklearn.metrics import plot_confusion_matrix
#plot_confusion_matrix(clf, X_test, y_test)





# κάθε φορά που θα τρέχουν οι αλγόριθμοι θα δίνουν διαφορετική ακρίβεια, καθώς κάθε φορά κάνουμε ξανά εκαπίδευση στο μοντέλο μας 
# τους βάλαμε να τρέξουν 10 φορές και συγκρίναμε τα αποτελέσματα των accuracy_score 
# έπειτα βγάλαμε κάποιες μεταβλητές και τρέξαμε τους άνω αλγόριθμους συγκεκριμένα κυρίως τα features με το μικρότερο συντελεστή συσχέτισης και ελέγξαμε πάλι
# την ακρίβεια τους. Εν τέλει καταλήξαμε στη χρήση αυτών των 9 πιο σχετικών feature (με το fried) dataset .
# Οι 3 πιο καλοί ήταν ο Logistic Regression , o Naive Bayes και ο Random Forest Trees 
# τα αποτελέσματα έχουν σημειωθεί σε 2 σετ με μεταβλητές 8 , 9 τω πλήθως αντίστοιχα και εφόσον έτρεξαν 10 φορές είδαμε ποιοι είναι πιο ακριβείς
# κάναμε ένα test ασυμπτωτικό μέσων τιμών για να συγκρίνουμε αν ο LR έχει καλύτερη μέση τιμή αποτελεσματικότητας από τον random forest trees και μετά
# ο καλύτερος ένα ακόμα test `με τον Νaive Bayes. Έγιναν στην R

#Τελικά δείχτηκε ότι ο logistic regfresion λειτουργεί καλύτερα ως μέθοδος classification στο project με τις μεταβλητές που πήραμε


# print(df.info())

# βρηκαμε accuracy στο logistic reggresion 0.731 το καλύτερο όλων , άρα κρατάμε τη συγκεκριμένη προσπάυεθεια με τον logreg 

with open('finalized_modellogreg1.sav' , 'rb') as f:
    mp = pickle.load(f)
    predict = mp.predict(xt)
    print(predict)
    result = mp.score(xt, yt)
    print(result)
    from sklearn.metrics import classification_report
    print(classification_report(yt,predict))
    from sklearn.metrics import plot_confusion_matrix
    plot_confusion_matrix(mp, xt, predict)
    
# θα κάνουμε και τρία pie charts , ένα για την κάθε κλάσση frail , prefrail και non frail αντίστοιχα με βάση το support που 
# δόθηκε από το classification report 
# εφόσον το μοντέλο μας είναι αποθηκευμένο (το logistic reggresion) και τα x test , y test είναι αποθηκευμένα ως csv (δεν θα ξακαναεκπαιδεύσουμε το συγκεκριμένο,
# -> δεν αλλάζουν τα σετ δεδομένων άρα θα βάλουμε απευθείας τα νούμερα από το support
# το βάζουμε σε σχόλια , διότι τρέχοντάς το όλο με jupyter μπλέκονται τα σχήματα


#slices = [48 ,47 , 13]
#labels = ["NON FRAIL" , "PRE-FRAIL" , "FRAIL"]
#colors = ("green" , "cyan" , "red")
#plt.pie(slices , labels = labels , colors = colors , wedgeprops = {"edgecolor" : "black"} , shadow = True ,startangle = 90 , autopct = '%1.1f%%'  )
#plt.title("PERCENTAGE OF FRIED CLASSIFICATION")
#plt.show()

# βλέπουμε ότι και στο confusion matrix 0 από το test set μας που ήταν frail δε βγήκαν frail ( τα πιο σημαντικά false negatives , άρα κρατάμε αυτή τη προσπάθεια 
    
#from sklearn.metrics import classification_report
#print(classification_report(y_test,y_pred))
#from sklearn.metrics import plot_confusion_matrix
#plot_confusion_matrix(logreg, X_test, y_test)



df_beac = pd.read_csv("/Users/evangelosgeraridis/Desktop/ddcdm/big data/PROJECT Μεγαλοοικονόμου/beacons_dataset.csv" , sep =';' )
    
# στα obvious erroneous values , θα κάνουμε τα εξής με το df.replace , όπως στο clinical

df_beac = df_beac.replace({ "Leavingroom" : "Livingroom" , "Livingroom1" : "Livingroom" , "Sitingroom" : "Livingroom" })

def Replace(i):
    try:
        float(i)
        return float(i)
    except:
           return np.nan

df_beac[['part_id']] = df_beac[['part_id']].applymap(func=Replace)  # οποιοδήποτε part_id δεν είναι numeric γίνεται nan ,καθώς είναι erroneous value
df_beac[['room']] = df_beac[['room']].replace({np.nan : ''})   # Τα nan τα κάνουμε εδώ empty string για να κάνουμε  drop τις στήλες όπου υπάρχει nan μετά , για να είναι μόνο στο part_id τα nan
df_beac = df_beac.dropna()

with pd.option_context('display.max_rows', None , 'display.max_columns', None):
    print(df_beac['room'].unique())
    

df_beac = df_beac.replace({"LeavingRoom" : "Livingroom","Leavingroom":"Livingroom",'Livroom':'Livingroom',"LivibgRoom" : "Livingroom","Livingroon" : "Livingroom","livingroom" : "Livingroom","SittingOver" : "Livingroom","Leavivinroom" : "Livingroom","Living" : "Livingroom" ,"Sittinroom" : "Livingroom","Liningroom" : "Livingroom", "Livingroom1" : "Livingroom" ,"Luvingroom1" : "Livingroom","LuvingRoom" : "Livingroom", "Sitingroom" : "Livingroom" ,"Sittingroom" : "Livingroom","SittingRoom" : "Livingroom","Livingroom2" : "Livingroom","LivingRoom2" : "Livingroom" ,"LivingRoom" : "Livingroom","SeatingRoom" : "Livingroom","Sittigroom" : "Livingroom","LuvingRoom" : "Livingroom",
                           "Kitcen" : "Kitchen",'Kitch':'Kitchen', "Kithen" : "Kitchen", "Kitchen2" : "Kitchen" , "Kitvhen" : "Kitchen", "kitchen" : "Kitchen", "Kichen" : "Kitchen" , "Kiychen" : "Kitchen", "Kitcheb" : "Kitchen",
                           "Desk" : "Office", "Workroom" : "Office", "Office1" : "Office", "Office2" : "Office", "Office1st" : "Office", "Office-2" : "Office",
                           "Washroom" : "Bathroom","Bathroon":"Bathroom",'Barhroom':'Bathroom',"Bathroom1" : "Bathroom" ,"Baghroom" : "Bathroom", "Bsthroom" : "Bathroom", "Bathroim" : "Bathroom", "Bqthroom" : "Bathroom", "Bathroom-1" : "Bathroom",
                           "DiningRoom" : "Dinningroom","Dinerroom": "Dinningroom","DinerRoom" : "Dinningroom", "DinnerRoom": "Dinningroom", "DinningRoom" : "Dinningroom",
                           "Bedroom2" : "Bedroom","Bed":"Bedroom", "Bedroom1" : "Bedroom","bedroom" : "Bedroom", "Bedroom1st" : "Bedroom", "Bedroom-1" : "Bedroom",
                           "Veranda" : "Garden", "Outdoor" : "Garden",
                           "Laundry" : "Laundryroom", "LaundryRoom" : "Laundryroom",
                           "Chambre" : "Chamber" , "Hall" : "Chamber","ExitHall" : "Chamber"})
print(df_beac['room'].unique())

def Replace(i):
    try:
        float(i)
        return float(i)
    except:
           return np.nan

df_beac[['part_id']] = df_beac[['part_id']].applymap(func=Replace)  # οποιοδήποτε part_id δεν είναι numeric γίνεται nan ,καθώς είναι erroneous value
df_beac[['room']] = df_beac[['room']].replace({np.nan : ''})   # Τα nan τα κάνουμε εδώ empty string για να κάνουμε  drop τις στήλες όπου υπάρχει nan μετά , για να είναι μόνο στο part_id τα nan
df_beac = df_beac.dropna()

with pd.option_context('display.max_rows', None , 'display.max_columns', None):
    print(df_beac.info())

print(df_beac['part_id'].unique())   # we have 124.


#κανουμε replace τα 3 ψηφια με nan
df_beac = df_beac.replace({123 : np.nan, 124 : np.nan})
df_beac['part_id'] = df_beac['part_id'].dropna()


df_beac = df_beac.sort_values(['part_id' , 'ts_date' , 'ts_time'])




from datetime import datetime, date, time, timedelta
df_beac['ts_time'] = pd.to_datetime(df_beac['ts_time'], format='%H:%M:%S')
df_beac['new_time']=(df_beac['ts_time'].dt.hour*60+df_beac['ts_time'].dt.minute)*60 + df_beac['ts_time'].dt.second

df_beac.head()

df_beac['time_diff'] = df_beac['new_time'] - df_beac['new_time'].shift()
df_beac['time_diff'] = df_beac['time_diff'].shift(-1).fillna(0)
df_beac.head()

df_beac['time_diff']=df_beac['time_diff'].abs()
df_beac= df_beac.drop(['ts_time','new_time'],axis=1)
df_beac.head()

df_beac['dup']=df_beac['ts_date'].duplicated()
df_beac.head(20)

df_beac['dup']=df_beac['dup'].replace({False:np.nan})
df_beac=df_beac.dropna()
df_beac=df_beac.drop('dup',axis=1)
df_beac.head()


#for i in range(46417):
 #   if df_beac['part_id'][i]==df_beac['part_id'][i+1] & df_beac['room'][i]==df_beac['room'][i+1]:
  #      df_beac['time_diff'][i]=df_beac['time_diff'][i]+df_beac['time_diff'][i+1]
#print(df_beac)

df_new=df_beac.groupby(['part_id','room']).sum()
df_new=df_new.drop('ts_date',axis=1)
df_new.head(20)

df_new2=df_beac.groupby('part_id').sum('time_diff')
df_new2=df_new2.drop('ts_date',axis=1)
df_new2 = df_new2.rename(columns={'time_diff': 'all_time'})
df_new2.head()


d1f=df_new.merge(df_new2,left_index=True,
right_index=True)



d1f['%time'] = (d1f.time_diff / d1f.all_time) * 100
d1f.head(20)


d1f = d1f.replace({ 'Garden': np.nan, 'TV': np.nan, 'Entry': np.nan, 'Right': np.nan,
'Left': np.nan, 'Entrance': np.nan, 'Chamber': np.nan, 'Office': np.nan, 'Two': np.nan, 'Three': np.nan, 'One': np.nan, 'Garage': np.nan,
'Dinningroom': np.nan, '2ndRoom': np.nan, 'Four': np.nan, 'three': np.nan, 'K' : np.nan,'T': np.nan, 'Box': np.nan, 'Guard': np.nan, 'Storage': np.nan,
'Pantry': np.nan, 'Library': np.nan, 'Laundryroom': np.nan, 'Box-1': np.nan})
d1f = d1f.dropna()
print(d1f.head())
d1f = d1f.drop(['time_diff'], axis = 1)
d1f = d1f.drop(['all_time'], axis = 1)


df_final = d1f.pivot_table('%time', ['part_id' ], 'room' )
df_final = df_final.replace({ np.nan : 0})
df_final=df_final.reset_index(level=['part_id'])
with pd.option_context('display.max_rows', None , 'display.max_columns', None):
    print(df_final)

#df_final=df_final.drop('',axis=1)
df_final.info()

#merge 2 datasets
df_last = pd.merge(df, df_final)
print(df_last.head(10))
print(df_last.info())


#smaller dataset
df_results=pd.merge(df[['part_id','activity_regular' , 'gait_speed_4m' , 'balance_single' , 'cognitive_total_score', 'depression_total_score' , 'age' , 'pain_perception' ,'health_rate' , 'leisure_out','fried']], df_final)
df_results = df_results.drop(['part_id'], axis=1)
print(df_results.head())

#pca for data visualization στο dataframe με τις 15 στήλες (χαρακτηριστικά)
# τα 9 που χρησιμοποιήθηκαν για classificaton + τα δωμάτια και fried 

# Εδώ κάνουμε pca και μετά kmeans , ωστόσο δοκιμάσαμε πρώτα kmeans μετά pca και μετά ξανά k means , όπως αναφέρουμε και στην αναφορά ,
# μαζί με τις οπτικοποιήσεις με διαγράμματα , ωστόσο λόγω της ακρίβειας που πέτυχε αποφασίσαμε να αφήσουμε στον κώδικα πρώτα pca και μετά kmeans

%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Apply Standardization to features matrix X
features = ['Bedroom','Bathroom','Kitchen','Livingroom','activity_regular' , 'gait_speed_4m' , 'balance_single' , 'cognitive_total_score', 'depression_total_score' , 'age' , 'pain_perception' ,'health_rate' , 'leisure_out']
x = df_results.loc[:, features].values
y = df_results.loc[:,['fried']].values

# Make an instance of PCA
pca = PCA(n_components=2)

# Fit and transform the data
principalComponents = pca.fit_transform(x)

principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])

finalDf = pd.concat([principalDf, df_results[['fried']]], axis = 1)

fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (8,8));
targets = df_results.loc[:, 'fried'].unique()
colors = ['r', 'g', 'b']

for target, color in zip(targets,colors):
    indicesToKeep = finalDf['fried'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)

ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)    
ax.legend(targets)
ax.grid()

#explained variance
#The explained variance tells us how much information (variance) can be attributed to each of the principal components.
pca.explained_variance_ratio_

sum(pca.explained_variance_ratio_)

#PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression

colnames = df_results.columns[:-1]

# Get all columns except the label column for the first image
values = df_results.loc[0, colnames].values

X_train, X_test, y_train, y_test = train_test_split(df_results[colnames], df_results['fried'], random_state=0)

#standardize the data
scaler = StandardScaler()

# Fit on training set only.
scaler.fit(X_train)

# Apply transform to both the training set and the test set.
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

"""
n_components = .90 means that scikit-learn will choose the minimum number 
of principal components such that 90% of the variance is retained.
"""

pca = PCA(n_components = .90)

# Fit PCA on training set only
pca.fit(X_train)

# Apply the mapping (transform) to both the training set and the test set. 
X_train = pca.transform(X_train)
X_test = pca.transform(X_test)

# Logistic Regression
clf = LogisticRegression()
clf.fit(X_train, y_train)

print('Number of dimensions before PCA: ' + str(len(colnames)))
print('Number of dimensions after PCA: ' + str(pca.n_components_))
print('Classification accuracy: ' + str(clf.score(X_test, y_test)))
#εδω βγαζει 4 features οτι μπορουν να αφαιρεθουν, αλλα δε βρηκα πως..

#KMEANS
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Generating the sample data from make_blobs
# This particular setting has one distinct cluster and 3 clusters placed close
# together.
X, y = make_blobs(
    n_samples=278,
    n_features=15,
    centers=4,
    cluster_std=1,
    center_box=(-10.0, 10.0),
    shuffle=True,
    random_state=1,
)  # For reproducibility

range_n_clusters = [2, 3, 4, 5, 6,7]

for n_clusters in range_n_clusters:
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(
        "For n_clusters =",
        n_clusters,
        "The average silhouette_score is :",
        silhouette_avg,
    )

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            ith_cluster_silhouette_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7,
        )

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(
        X[:, 0], X[:, 1], marker=".", s=30, lw=0, alpha=0.7, c=colors, edgecolor="k"
    )

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(
        centers[:, 0],
        centers[:, 1],
        marker="o",
        c="white",
        alpha=1,
        s=200,
        edgecolor="k",
    )

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker="$%d$" % i, alpha=1, s=50, edgecolor="k")

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(
        "Silhouette analysis for KMeans clustering on sample data with n_clusters = %d"
        % n_clusters,
        fontsize=14,
        fontweight="bold",
    )

plt.show()

import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans

data = df_results
x = data.iloc[:,1:] # 1t for rows and second for columns

kmeans = KMeans(4)
kmeans.fit(x)

identified_clusters = kmeans.fit_predict(x)
identified_clusters

data_with_clusters = data.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['fried'],data_with_clusters['Kitchen'],c=data_with_clusters['Clusters'],cmap='rainbow')

data['cluster'] = identified_clusters
with pd.option_context('display.max_rows', None , 'display.max_columns', None):
    print(data)
    
plt.scatter(data['fried'] , data['cluster'])
fig = plt.figure(figsize =(10, 7))
plt.show()

ncl0 = (data['cluster'] == 0).sum()
ncl1 = (data['cluster'] == 1).sum()
ncl2 = (data['cluster'] == 2).sum()
ncl3 = (data['cluster'] == 3).sum()
ncleach = [ncl0, ncl1 , ncl2 , ncl3]

# of people in cluster

clusterss = [0,1,2,3]
plt.barh(clusterss , ncleach)
plt.title(" Number of people in each cluster")
plt.ylabel("cluster")
plt.xlabel("number of people")
print(ncleach)

#k=4
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans


z = data.iloc[:,1:] # 1t for rows and second for columns

d1f.head(10)


# Τώρα θα δοκιμάσουμε την ίδια διαδικασία στο preprocesed dataset (με 44 στήλες ) , δεν βάλαμε το τελείως αρχικό clinical dataset  , στο merge
# καθώς είχαμε πρόβλημα με τα missing values κάποιων features (ήταν πολλά για να γεμίσουμε με median grouped by fried)

data2 = df_last.drop(['part_id'], axis=1)
X = data2
#pca for data visualization
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Το ίδιο έγινε και εδώ κρατήσαμε pca και μετά k means

# Apply Standardization to features matrix X
features = ['fried','gender','age','hospitalization_one_year','ortho_hypotension','vision','audition','raise_chair_time' ,'balance_single','gait_speed_4m','gait_optional_binary','falls_one_year','fractures_three_years' ,'bmi_score','screening_score','cognitive_total_score' ,'memory_complain' ,'sleep','depression_total_score' ,'anxiety_perception'  ,'living_alone', 'leisure_out'  ,'leisure_club','social_visits' ,'social_calls'  ,'social_phone' , 'social_skype' , 'social_text' ,'life_quality' ,'health_rate' ,'health_rate_comparison' ,'pain_perception' ,'activity_regular' ,'smoking' , 'alcohol_units'  ,'katz_index',  'iadl_grade','comorbidities_count', 'comorbidities_significant_count',  'Bathroom','Bedroom','Kitchen'  ,'Livingroom']

x = data2.loc[:, features].values
y = data2.loc[:,['fried']].values

# Make an instance of PCA
pca = PCA(n_components=2)

# Fit and transform the data
principalComponents = pca.fit_transform(X)

principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])

finalDf = pd.concat([principalDf, data2[['fried']]], axis = 1)

fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (8,8));
targets = data2.loc[:, 'fried'].unique()
colors = ['r', 'g', 'b']

for target, color in zip(targets,colors):
    indicesToKeep = finalDf['fried'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)

ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)    
ax.legend(targets)
ax.grid()

#explained variance
#The explained variance tells us how much information (variance) can be attributed to each of the principal components.
pca.explained_variance_ratio_

sum(pca.explained_variance_ratio_)

#PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression

colnames = data2.columns[:-1]

# Get all columns except the label column for the first image
values = data2.loc[0, colnames].values

X_train, X_test, y_train, y_test = train_test_split(data2[colnames], data2['fried'], random_state=0)

#standardize the data
scaler = StandardScaler()

# Fit on training set only.
scaler.fit(X_train)

# Apply transform to both the training set and the test set.
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

"""
n_components = .90 means that scikit-learn will choose the minimum number 
of principal components such that 90% of the variance is retained.
"""

pca = PCA(n_components = .90)

# Fit PCA on training set only
pca.fit(X_train)

# Apply the mapping (transform) to both the training set and the test set. 
X_train = pca.transform(X_train)
X_test = pca.transform(X_test)

# Logistic Regression
clf = LogisticRegression()
clf.fit(X_train, y_train)

print('Number of dimensions before PCA: ' + str(len(colnames)))
print('Number of dimensions after PCA: ' + str(pca.n_components_))
print('Classification accuracy: ' + str(clf.score(X_test, y_test)))
X, y = make_blobs(
    n_samples=278,
    n_features=44,
    centers=4,
    cluster_std=1,
    center_box=(-10.0, 10.0),
    shuffle=True,
    random_state=1,
)  # For reproducibility

range_n_clusters = [2, 3, 4, 5, 6,7]

for n_clusters in range_n_clusters:
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(
        "For n_clusters =",
        n_clusters,
        "The average silhouette_score is :",
        silhouette_avg,
    )

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            ith_cluster_silhouette_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7,
        )

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(
        X[:, 0], X[:, 1], marker=".", s=30, lw=0, alpha=0.7, c=colors, edgecolor="k"
    )

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(
        centers[:, 0],
        centers[:, 1],
        marker="o",
        c="white",
        alpha=1,
        s=200,
        edgecolor="k",
    )

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker="$%d$" % i, alpha=1, s=50, edgecolor="k")

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(
        "Silhouette analysis for KMeans clustering on sample data with n_clusters = %d"
        % n_clusters,
        fontsize=14,
        fontweight="bold",
    )

plt.show()

import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans


plt.scatter(data2['fried'],data2['Kitchen'])
plt.show()

x = data2.iloc[:,1:] # 1t for rows and second for columns

# και με όλο το preprocesed dataframe είχαμε με k means για 4 cluster highest shillouette score
# για όλο το dataframe λοιπόν έχουμε :

# όπως φαίνεται και στην αναφορά μας , δοκιμάσαμε με όλο το πιθανό πλήθος clusters , αλλά και πάλι το καλύτερο αποτέλεσμα με 4 clusters
# although since we know we have to expect 3 clusters from the datasets we will avisualise with 3 clusters

kmeans = KMeans(3)
kmeans.fit(x)

identified_clusters = kmeans.fit_predict(x)
identified_clusters

data2_with_clusters = data2.copy()
data2_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data2_with_clusters['fried'],data2_with_clusters['Kitchen'],c=data2_with_clusters['Clusters'],cmap='rainbow')


# VISUALIZATIONS

data2['cluster'] = identified_clusters

mcl0 = (data2['cluster'] == 0).sum()
mcl1 = (data2['cluster'] == 1).sum()
mcl2 = (data2['cluster'] == 2).sum()
mcleach = [mcl0, mcl1 , mcl2]

# of people in cluster

identified_clusters = kmeans.fit_predict(x)
clusterzs = [0,1,2]
plt.barh(clusterzs , mcleach)
plt.title(" Number of people in each cluster")
plt.ylabel("cluster")
plt.xlabel("number of people")
print(mcleach)
plt.show()


fried_in_zero_clusterr = (data2['cluster'] == 0).groupby(data2['fried']).sum()
print(fried_in_zero_clusterr)
fried_in_first_clusterr = (data2['cluster'] == 1).groupby(data2['fried']).sum()
print(fried_in_first_clusterr)
fried_in_second_clusterr = (data2['cluster'] == 2).groupby(data2['fried']).sum()
print(fried_in_second_clusterr)

slices0 = [fried_in_zero_clusterr[0] ,fried_in_zero_clusterr[1] , fried_in_zero_clusterr[2]]
labels = ["NON FRAIL" , "PRE-FRAIL" , "FRAIL"]
colors = ("lime" , "cyan" , "purple")
plt.pie(slices0 , labels = labels , colors = colors , wedgeprops = {"edgecolor" : "black"} , shadow = True ,startangle = 90 , autopct = '%1.1f%%' )
plt.title("FRIED CATEGORIES IN CLUSTER NUMBER 0")
plt.show()

slices1 = [fried_in_first_clusterr[0] ,fried_in_first_clusterr[1] , fried_in_first_clusterr[2]]
labels = ["NON FRAIL" , "PRE-FRAIL" , "FRAIL"]
colors = ("lime" , "cyan" , "purple")
plt.pie(slices1 , labels = labels , colors = colors , wedgeprops = {"edgecolor" : "black"} , shadow = True ,startangle = 90 , autopct = '%1.1f%%' )
plt.title("FRIED CATEGORIES IN CLUSTER NUMBER 1")
plt.show()

slices2 = [fried_in_second_clusterr[0] ,fried_in_second_clusterr[1] , fried_in_second_clusterr[2]]
labels = ["NON FRAIL" , "PRE-FRAIL" , "FRAIL"]
colors = ("lime" , "cyan" , "purple")
plt.pie(slices2 , labels = labels , colors = colors , wedgeprops = {"edgecolor" : "black"} , shadow = True ,startangle = 90 , autopct = '%1.1f%%' )
plt.title("FRIED CATEGORIES IN CLUSTER NUMBER 2")
plt.show()


# check if there is visually a correlation between elders spending more time in the living room for example and their fraility

plt.scatter(data['Bathroom'],data['fried'])
plt.show()
plt.scatter(data2['Bathroom'],data2['fried'])
plt.show()

plt.scatter(data['Bedroom'],data['fried'])
plt.show()
plt.scatter(data2['Bedroom'],data2['fried'])
plt.show()


plt.scatter(data2['Livingroom'],data2['fried'])
plt.show()

plt.scatter(data['Kitchen'],data['fried'])
plt.show()
plt.scatter(data2['Kitchen'],data2['fried'])
plt.show()


fried_in_kitchen = ((data2['Kitchen']).groupby(data2['fried']).sum())
fried_in_bathroom = ((data2['Bathroom']).groupby(data2['fried']).sum())
fried_in_bedroom = ((data2['Bedroom']).groupby(data2['fried']).sum())
fried_in_livingroom = ((data2['Livingroom']).groupby(data2['fried']).sum())
# finding percentage of time spent by elders in each room , given that their fried category is pre frail

slicespf = [fried_in_kitchen[1]/(fried_in_kitchen[1] + fried_in_bathroom[1] +fried_in_bedroom[1] + fried_in_livingroom[1]) ,fried_in_bathroom[1]/(fried_in_kitchen[1] + fried_in_bathroom[1] +fried_in_bedroom[1] + fried_in_livingroom[1]) , fried_in_bedroom[1]/(fried_in_kitchen[1] + fried_in_bathroom[1] +fried_in_bedroom[1] + fried_in_livingroom[1]) , fried_in_livingroom[1]/(fried_in_kitchen[1] + fried_in_bathroom[1] +fried_in_bedroom[1] + fried_in_livingroom[1])  ]
labels = ['KITCHEN' , 'BATHROOM' , 'BEDROOM' , 'LIVING ROOM']
colors = ("pink" , "cyan" , "red" , "yellow")
plt.pie(slicespf , labels = labels , colors = colors , wedgeprops = {"edgecolor" : "black"} , shadow = True ,startangle = 90 , autopct = '%1.1f%%'  )
plt.title("PERCENTAGE OF TIME SPENT IN EACH ROOM BY PRE - FRAIL ELDERS")
plt.show()

slicesnf = [fried_in_kitchen[0]/(fried_in_kitchen[0] + fried_in_bathroom[0] +fried_in_bedroom[0] + fried_in_livingroom[0]) ,fried_in_bathroom[0]/(fried_in_kitchen[0] + fried_in_bathroom[0] +fried_in_bedroom[0] + fried_in_livingroom[0]) , fried_in_bedroom[0]/(fried_in_kitchen[0] + fried_in_bathroom[0] +fried_in_bedroom[0] + fried_in_livingroom[0]) , fried_in_livingroom[0]/(fried_in_kitchen[0] + fried_in_bathroom[0] +fried_in_bedroom[0] + fried_in_livingroom[0])  ]
labels = ['KITCHEN' , 'BATHROOM' , 'BEDROOM' , 'LIVING ROOM']
colors = ("pink" , "cyan" , "red" , "yellow")
plt.pie(slicesnf , labels = labels , colors = colors , wedgeprops = {"edgecolor" : "black"} , shadow = True ,startangle = 90 , autopct = '%1.1f%%'  )
plt.title("PERCENTAGE OF TIME SPENT IN EACH ROOM BY NON FRAIL ELDERS")
plt.show()

slicesf = [fried_in_kitchen[2]/(fried_in_kitchen[2] + fried_in_bathroom[2] +fried_in_bedroom[2] + fried_in_livingroom[2]) ,fried_in_bathroom[2]/(fried_in_kitchen[0] + fried_in_bathroom[2] +fried_in_bedroom[2] + fried_in_livingroom[2]) , fried_in_bedroom[2]/(fried_in_kitchen[2] + fried_in_bathroom[2] +fried_in_bedroom[2] + fried_in_livingroom[2]) , fried_in_livingroom[2]/(fried_in_kitchen[2] + fried_in_bathroom[2] +fried_in_bedroom[2] + fried_in_livingroom[2])  ]
labels = ['KITCHEN' , 'BATHROOM' , 'BEDROOM' , 'LIVING ROOM']
colors = ("pink" , "cyan" , "red" , "yellow")
plt.pie(slicesf , labels = labels , colors = colors , wedgeprops = {"edgecolor" : "black"} , shadow = True ,startangle = 90 , autopct = '%1.1f%%'  )
plt.title("PERCENTAGE OF TIME SPENT IN EACH ROOM BY FRAIL ELDERS")
plt.show()


# classification pie chart (shows percentage of fried categories in classes

# αναφερόμαστε στο σταθερό μοντέλο και εφόσον έχουν αποθηκευτεί σε csv τα xtest και y test του μοντέλου μας ,
# δεν αλλάζει το πλήθος των frail για παράδειγμα σε κάθε class

slicesc = [48 ,47 , 13]
labels = ["NON FRAIL" , "PRE-FRAIL" , "FRAIL"]
colors = ("pink" , "cyan" , "red")
plt.pie(slicesc , labels = labels , colors = colors , wedgeprops = {"edgecolor" : "black"} , shadow = True ,startangle = 90 , autopct = '%1.1f%%'  )
plt.title("PERCENTAGE OF FRIED CLASSIFICATION")
plt.show()

# PERCENTAGE OF FRIED CATEGORIES IN THE WHOLE DATASET
df_mh_peiragmeno_clinical = pd.read_csv('/Users/evangelosgeraridis/Desktop/ddcdm/big data/PROJECT Μεγαλοοικονόμου/clinical_dataset.csv', sep=';')
nfrail = (df_mh_peiragmeno_clinical['fried'] == 'Frail').sum()
nprefrail = (df_mh_peiragmeno_clinical['fried'] == 'Pre-frail').sum()
nnonfrail = (df_mh_peiragmeno_clinical['fried'] == 'Non frail').sum()
slicesod = [nfrail , nprefrail , nnonfrail]
labels = ["NON FRAIL" , "PRE-FRAIL" , "FRAIL"]
colors = ("pink" , "cyan" , "red")
plt.pie(slicesod , labels = labels , colors = colors , wedgeprops = {"edgecolor" : "black"} , shadow = True ,startangle = 90 , autopct = '%1.1f%%'  )
plt.title("PERCENTAGE OF FRIED CATEGORIES IN ORIGINAL DATASET")
plt.show()





```
